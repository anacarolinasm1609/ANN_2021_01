```python
import zipfile
file_name = zipfile.ZipFile('trash.zip', 'r')
file_name.extractall()
```

```python
# all images are load in variable x...
FILE='trash.zip'
img_name=[]
x=[]
with zipfile.ZipFile(FILE, "r") as f:
    for name in f.namelist():
        img_name.append(name)
        x.append(cv2.imread(name))
        #x.append(cv2.resize(cv2.imread(name),(64,64)))
```

```python
Ind = 6
img=plt.imread(img_name[Ind])
plt.axis("off")
plt.imshow(x[Ind])
plt.title(img_name[Ind])
```

```python
y_names = []
for x in img_name:
  name = x[0]+x[1]+x[2]
  y_names.append(name)
print(y_names)
```

```python
unique_list = []
     
for x in y_names:
  # check if exists in unique_list or not
  if x not in unique_list:
    unique_list.append(x)

print(unique_list)
```

```python
n = len(unique_list)
y = []
for x in y_names:
  for i in range(n):
    if(x == unique_list[i]):
      y.append(i)
print(y)
```
